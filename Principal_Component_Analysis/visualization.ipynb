```python
import random

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from PCA import PCA
```

```python
# Load Data
data_for_classification = pd.read_csv('../data/for_classification.csv')
data_for_classification.head()
```

```python
# Principal Component Analysis(n_components=2)
x = data_for_classification.iloc[:, 1:4].to_numpy()
pca = PCA(n_components=2)
x_reduced = pca.fit_transform(x)
```

```python
# Visualization
labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['crimson', 'deepskyblue', 'mediumpurple']

plt.style.use('ggplot')
fig, ax = plt.subplots()
fig.set_figheight(7.2)
fig.set_figwidth(10.8)
for color, label in zip(colors, labels):
    idx = list(map(lambda x: x-1, data_for_classification.loc[data_for_classification['Species'] == label]['Id']))
    ax.scatter(x_reduced[idx, 0], x_reduced[idx, 1], c=color, label=label)
plt.title('Principal Component Analysis with Iris Dataset')

ax.legend()
ax.grid(True)

plt.show()
```
